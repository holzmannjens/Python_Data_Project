```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## JOB COUNTS


```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median') #until here I qould also just run the code with the groupby function (df.groupby('job_title_short')['salary_year_avg'].median())
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median') #but here I broke it down to the different job_countries (grupby method lost here)
```

```python
top_countries = df['job_country'].value_counts().head(6).index
df_top_country_salary = df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
df_top_country_salary = df_top_country_salary.loc[top_countries]
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_top_country_salary = df_top_country_salary[job_titles]
df_top_country_salary.plot(kind='bar')
plt.ylabel('Median Salary($USD)')
plt.xlabel('')
plt.title('Median Salary by Job Title and Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```
